```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df
```

```python
df.head()
```

```python
sns.lineplot(x='tip', y='total_bill',data=df)
```

```python
sns.countplot(x="smoker",data=df) # to count the data
```

```python
sns.countplot(x="sex",data=df)
```

```python
sns.pairplot(data=df,hue="tip") #  hue is for selecting color, tip is the column which is selected
```

```python
sns.scatterplot(data=df)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
```

```python
sns.histplot(data=df)
```

```python
sns.histplot(x="total_bill", y="tip", data=df)
```

```python
sns.histplot(x="size", data=df)
```
